##**working with polars**

```python
import polars as pl
import time

start = time.time()
df = pl.scan_csv('drive/My Drive/yab/flight_csv_file/flights.csv').filter(
        (pl.col('MONTH') == 12) &(pl.col('ORIGIN_AIRPORT') == 'SEA') &
        (pl.col('DESTINATION_AIRPORT') == 'DFW')).collect()
end = time.time()
print(end - start)
display(df)
```
